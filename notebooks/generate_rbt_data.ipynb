```python
import pandas as pd
import numpy as np
import random
```

```python
# Set seed for reproducibility
np.random.seed(42)
random.seed(42)
```

```python
# Number of RBTs in the dataset
num_rbt = 3000
```

```python
# Generate synthetic data
rbt_data = {
    "rbt_id": [f"RBT_{i+1}" for i in range(num_rbt)],
    "age": np.random.randint(20, 50, num_rbt),
    "education_level": np.random.choice(["High School", "Bachelor's", "Master's"], num_rbt, p=[0.5, 0.4, 0.1]),
    "years_experience": np.round(np.random.uniform(0.5, 10, num_rbt), 1),
    "company_tenure": np.round(np.random.uniform(0.2, 7, num_rbt), 1),
    "hourly_wage": np.random.randint(15, 30, num_rbt),
    "caseload_size": np.random.randint(3, 12, num_rbt),
    "weekly_hours": np.random.randint(20, 40, num_rbt),
    "travel_time_per_week": np.random.randint(0, 10, num_rbt),
    "supervision_hours_last_month": np.random.randint(1, 10, num_rbt),
    "bcba_feedback_score": np.random.randint(1, 6, num_rbt),
    "rbt_support_rating": np.random.randint(1, 6, num_rbt),
    "supervision_helpfulness_score": np.random.randint(1, 6, num_rbt),
    "bcba_communication_score": np.random.randint(1, 6, num_rbt),
    "bcba_goal_communication_score": np.random.randint(1, 6, num_rbt),
    "supervision_frequency_last_3_months": np.random.randint(1, 15, num_rbt),
    "burnout_score": np.random.randint(1, 10, num_rbt),
    "work_life_balance": np.random.randint(1, 6, num_rbt),
    "turnover_status": np.random.choice([0, 1], num_rbt, p=[0.7, 0.3])  # 30% turnover rate
}
```

```python
# Create DataFrame
df_rbt = pd.DataFrame(rbt_data)
```

```python
# Save dataset to CSV
csv_filename = "data/rbt_turnover_data.csv"
df_rbt.to_csv(csv_filename, index=False)
```
